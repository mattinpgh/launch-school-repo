## Question 2
How can you determine whether a given string ends with an exclamation mark (!)? Write some code that prints True or False depending on whether the string ends with an exclamation mark.

````python
def exclaims(str):
    return True if str[-1] == "!" else False

str1 = "Come over here!"  # True
str2 = "What's up, Doc?"  # False
print(exclaims(str1))
print(exclaims(str2))
````

The suggested solution uses the string method endswith(), which is handy. I should remember it - it has a start and end value, which should be useful for file extensions


## Question 3
Starting with the string `famous_words = "seven years ago..."` show two different ways to create a new string with "Four score and" prepended to the front of the string.
